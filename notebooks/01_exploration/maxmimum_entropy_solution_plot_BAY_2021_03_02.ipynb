```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
dz = 0.01
zz = np.arange(0,1,dz)

#boltzmann gibbs

#parameter vectors:
c = np.arange(0.01,2,0.01)**3

for ci in c:
    u = np.log(zz+ci) - np.log(ci)
    ubar = np.trapz(u)*dz
    u /= ubar
    plt.plot(u,zz)
```

```python
#Tsallis: 

q = 2

c = np.arange(0.01,2,0.01)

for ci in c:
    u = (zz+ci)**((q-1)/q) - ci**((q-1)/q)
    ubar = np.trapz(u)*dz
    u /= ubar
    plt.plot(u,zz)
```
