```python
from sympy import *
z, c, q, a, q0 = symbols('z c q a q0')
```

```python

u = (z+c)**a - c**a
u_bar = integrate(u,(z,0,1),conds='none')
u = u/u_bar
chi_2 = integrate(u**2,(z,0,1),conds='none')
chi_2_q = chi_2
print(latex(simplify(chi_2.subs(a,(q-1)/q)   ) ) ) 
q = 1.5
chi_2 = simplify(chi_2.subs(a,(q-1)/q))
pprint(chi_2)
print(latex(chi_2))
```
