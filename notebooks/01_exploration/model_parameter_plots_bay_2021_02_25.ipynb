```python
from sympy import *
```

```python
a, r, z, chi = symbols('a r z chi')
u = (r+1)*(1-(1-z)**r)/r
```

```python
chi_2 = integrate(u**2,(z,0,1),conds='none')
chi_2 = simplify(chi_2)
pprint(chi_2)
```

```python
r_mod = 0.5*(2-chi)/(chi-1)
plot(r_mod,(chi,1.01,1.5))
```

```python
#energy diffusion:
ediff = integrate(u.diff(z)**2,(z,0,1))
#ediff = ((r+1)**2)/(r**4*(2*r-1))
ediff = ediff.subs(r,r_mod)
ediff = simplify(ediff)
pprint(ediff)
plot(ediff,(chi,1.01,1.3))
```
